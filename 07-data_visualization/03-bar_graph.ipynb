```python
plt.bar([0,1,2],[10,20,15]) # Current year
plt.bar([0,1,2],[20,10,12]) # Next year

plt.show()
```

```python
x_coordinates = np.array([0,1,2])

plt.bar(x_coordinates,[10,20,15]) # Current year
plt.bar(x_coordinates + 0.5,[20,10,12]) # Next year

plt.show()
```

```python
x_coordinates = np.array([0,1,2])

plt.bar(x_coordinates,[10,20,15], width = 0.5) # Current year
plt.bar(x_coordinates + 0.5,[20,10,12], width = 0.5) # Next year

plt.show()
```

```python
x_coordinates = np.array([0,1,2])*2

plt.bar(x_coordinates,[10,20,15], width = 0.5,label="Current Year") # Current year
plt.bar(x_coordinates + 0.5,[20,10,12], width = 0.5,label="Next Year") # Next year
plt.legend()
plt.show()
```

```python
x_coordinates = np.array([0,1,2])*2

plt.bar(x_coordinates-0.25,[10,20,15], width = 0.5,label="Current Year",tick_label=["Gold", "Platinum", "Silver"]) # Current year
plt.bar(x_coordinates + 0.25,[20,10,12], width = 0.5,label="Next Year") # Next year
plt.title("Metals Price Comparison")
plt.xlabel("Metal")
plt.ylabel("Price")

plt.legend()
plt.show()
```

```python
plt.style.use("dark_background")
x_coordinates = np.array([0,1,2])*2

plt.bar(x_coordinates-0.25,[10,20,15], width = 0.5,label="Current Year",tick_label=["Gold", "Platinum", "Silver"], color="green") # Current year
plt.bar(x_coordinates + 0.25,[20,10,12], width = 0.5,label="Next Year",color="orange") # Next year
plt.title("Metals Price Comparison")
plt.xlabel("Metal")
plt.ylabel("Price")

plt.legend()
plt.show()
```

```python
plt.style.use("dark_background")
x_coordinates = np.array([0,1,2])*2

plt.bar(x_coordinates-0.25,[10,20,15], width = 0.5,label="Current Year",tick_label=["Gold", "Platinum", "Silver"], color="yellow") # Current year
plt.bar(x_coordinates + 0.25,[20,10,12], width = 0.5,label="Next Year",color="orange") # Next year
plt.title("Metals Price Comparison")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.ylim(0,40)
plt.xlim(-2,5)
plt.legend()
plt.show()
```
