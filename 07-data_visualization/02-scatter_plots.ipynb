```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Line Plot
x = np.arange(10)
y1 = x**2
y2 = 2*x + 3
print(x)
print(y1)
print(y2)
```

```python
plt.style.use("seaborn")
```

```python
plt.figure(figsize=(10,10))   # Adjust the size of any plot
plt.scatter(x,y1,color="red", label="Apple",marker="^")
plt.scatter(x,y2, color="green", label="Kiwi",linestyle="dashed")
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Scatter Plot - Prices Over Time")
plt.legend()

plt.show()
```
