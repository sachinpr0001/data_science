## Generate a Normal Distribution (1-D Distribution)

```python
import numpy as np
import matplotlib.pyplot as plt
u = 0
sigma = 1
# Such a dist. (with mean = 0 and sigma = 1, is also called Standard Normal Distribution)

vals = np.random.randn(1000)

print(vals.shape)
```

```python
# now, changing the mean.

u = 5
sigma = 2

vals = u + sigma*np.random.randn(1000)

# 1000 = number of points i want to have

print(vals.shape)
# why did we do this????
# - ans in notebook
plt.hist(vals,50)

# 50 = number of buckets for the histogram
# if it were 10, witdth of the bar (bucket) would increase bcoz more points will go in same bucket
plt.show()
```

```python
# now, changing the sigma back to 1 to see the that sigma defines the spread

u = 5
sigma = 1

vals = u + sigma*np.random.randn(1000)

print(vals.shape)
# why did we do this????
# - ans in notebook
plt.hist(vals,50)
plt.show()
```

```python
# What is this histogram depicting?
# - it is showing how many values(no of values OR probability of X) fall in each intervals.
```

```python
vals = np.round(vals)
z = np.unique(vals, return_counts=True)
print(z)
```

```python
# first array --> numbers
# second array --> frequency (numbers closer to each number from first array)
```

## Multivariate Normal Distribution

```python
# Bivariate

mean = np.array([0.0, 0.0])
cov = np.array([[1,0],[0,1]])

dist = np.random.multivariate_normal(mean,cov, 500) # 500 = no. of points

plt.scatter(dist[:,0],dist[:,1])
```

```python
# mean obtained in the scatter above is 0.
#sigma(xx) = 1
#sigma(yy) = 1 (in case x is height and y is weight)
#sigma(xy) = sigma(yx) = 0  (in case x axis is height, y axis is colour, there is no relation)
```

```python
# Bivariate

mean = np.array([0.0, 0.0])
cov = np.array([[1,0.5],[0.5,1]])

dist = np.random.multivariate_normal(mean,cov, 500) # 500 = no. of points

plt.scatter(dist[:,0],dist[:,1])
```

```python
# change the covarience from 0 to 0.5 (giving a more positive value), it got tilted
```

```python
# Bivariate

mean = np.array([0.0, 0.0])
cov = np.array([[1,0.8],[0.8,1]])

dist = np.random.multivariate_normal(mean,cov, 500) # 500 = no. of points

plt.scatter(dist[:,0],dist[:,1])
```

```python
# changing it to 0.8, it gets more tilted.
```

```python
# one more distribution

mean = np.array([0.0, 0.0])
cov = np.array([[1,0.8],[0.8,1]])

mean2 = np.array([5.0, 6.0])
cov2 = np.array([[1.3,0.2],[0.2,1.1]])

dist = np.random.multivariate_normal(mean,cov, 500) # 500 = no. of points
dist2 = np.random.multivariate_normal(mean2,cov2, 500)

plt.scatter(dist[:,0],dist[:,1])
plt.scatter(dist2[:,0],dist2[:,1])
plt.show()
```

```python
# spread along y axis is more in orange than in blue.
# in blue, sigma is 1 and in orange, sigma is 1.1
# mean for blue = 0
# mean for orange = 5.5
```
