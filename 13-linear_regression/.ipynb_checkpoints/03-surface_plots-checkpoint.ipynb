### Surface Plots | Data Visualization
Surface Plots are used to
- Visualize Loss Functions in ML and Deep Learning
- Visualize State or State Value Functions in Reinforcement Learning

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#a = np.array([1,2,3])
#b = np.array([4,5,6,7])
a = np.arange(-1,1,0.02)
b = a
a,b = np.meshgrid(a,b)
#print(a)
#print(b)
```

```python
fig = plt.figure()
axes = fig.gca(projection = '3d')
axes.plot_surface(a, b, a**2+b**2, cmap = 'rainbow') #cmap = 'coolwarm'
plt.show()
```

### Contour Plots

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.contour(a,b,a**2+b**2, cmap ='rainbow')
plt.title('Contour Plot')
plt.show()
```
