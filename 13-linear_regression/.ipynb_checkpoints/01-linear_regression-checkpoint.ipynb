```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
Y = (X-5)**2

print(X,Y)
```

## GOAL
#### Given a function f(x), we want to find the value of x that minimizes f.

### Visualization

```python
plt.style.use('seaborn')
plt.plot(X,Y)
plt.ylabel("F(X)")
plt.xlabel("X")
plt.show()
```

```python
x = 0
#lr = learning rate
lr = 0.1
#50 steps in the downhill direction
for i in range(50):
    grad = 2*(x-5) # differentiation of Y = (X-5)**2 written directly
    x = x - lr*grad
    print(x)
```

```python
# from the 50 results above, we can see that the answers are almost 5
# ideally, it should be 5
# and it confirms that the algorithm for the function (or the curve) 
# is converging.
```

```python
#let's say f(x) denotes error. our goal is to minimize error.
# store the value of error at each iteration
x = 0
#lr = learning rate
lr = 0.1
error = [] # the list that stores arror at each iteration
#50 steps in the downhill direction
plt.plot(X,Y)
for i in range(50):
    grad = 2*(x-5) # differentiation of Y = (X-5)**2 written directly
    x = x - lr*grad
    # since, error (e) is f(x) here, e = y
    y = (x-5)**2
    error.append(y)
    plt.scatter(x,y)
    print(x)
```

```python
# Effective Step Size explained in written (notebook)
```

```python
# Plot the values of error
plt.plot(error)
plt.show()
```

```python
# as we increase the number of steps, the error reduces continuously.
# after reaching 20 steps (when 5 is reached in value), it saturates.
# bcoz at 5, the gradient becomes 0.
# we can put an if condition.
```
